```python
from random import randint


print('Let\'s play Rock, Paper, Scissors! Can you beat the Computer?\n')
print('How to Play:\nEnter Rock, Paper, or Scissors to declare your move\nEnter \'Done\' when finished playing!\nHave Fun!')
print('-----------------------------------------------------------')

#Create a list of play options
t = ["Rock", "Paper", "Scissors"]

#Assign a random play to the computer
computer = t[randint(0,2)]

#Set player to False
player = False

# Create accumulator for wins, losses ties, and rounds
win = 0
loss = 0
tie = 0
r = 0

while player == False:
#set player to True
    player = input("Your Move: ")
    print('')
    if player == 'Done':
        print('Game Finished! You won', win ,'time(s), lost', loss, 'time(s) and tied', tie, 'time(s)!')
        break
    print('Computer Move: ', computer, '\n')
    if player == computer:
        r += 1
        tie += 1
        print("Round", r, ": Tie!")
        print('-----------------------------------------------------')

    elif player == "Rock":
        if computer == "Paper":
            r += 1
            print("Round", r, ": You lose!", computer, "covers", player)
            print('-----------------------------------------------------')
            loss += 1
        else:
            r += 1
            print("Round", r, ": You win!", player, "smashes", computer)
            print('-----------------------------------------------------')

            win += 1
    elif player == "Paper":
        if computer == "Scissors":
            r += 1
            print("Round", r, ": You lose!", computer, "cut", player)
            print('-----------------------------------------------------')
            loss += 1
        else:
            r += 1
            print("Round", r, ": You win!", player, "covers", computer)
            print('-----------------------------------------------------')
            win += 1
    elif player == "Scissors":
        if computer == "Rock":
            r += 1
            print("Round", r, ": You lose...", computer, "smashes", player)
            print('-----------------------------------------------------')
            loss += 1
        else:
            r += 1
            print("Round", r, ": You win!", player, "cut", computer)
            print('-----------------------------------------------------')
            win += 1
    #elif player == 'Done':
        #break
    else:
        print("That's not a valid play. Check your spelling!")
    #player was set to True, but we want it to be False so the loop continues
    player = False
    computer = t[randint(0,2)]
```
